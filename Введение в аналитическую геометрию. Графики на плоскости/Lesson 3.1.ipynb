**1. Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)**


```python
import numpy as np
```

Для нахождения суммы векторов, надо сложить их соответствующие координаты

```python
a = np.array ([10,10,10])
b = np.array ([0,0,-10])
c = a+b
c
```

```python
np.sqrt(c[0]**2+c[1]**2+c[2]**2) # расчет длины вектора с
```

```python
def length_vec(vec): # функция для расчета длины вектора в трехмерном пространстве
    return np.sqrt(vec[0]**2+vec[1]**2+vec[2]**2)
```

```python
length_vec(c)
```

```python
length_vec(a)
```

```python
length_vec(b)
```

**2. Задание.
Почему прямые не кажутся перпендикулярными?**


```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y2)
# сделаем одинаковый масштаб осей х и у, и тогда прямые станут визуально перпендикулярными
plt.xlim(-5, 5) 
plt.ylim(-5, 5)
plt.xlabel("x")
plt.ylabel("y")
```

**3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.**


```python
R = 5
x = np.linspace(-2*R, 2*R, 8000)
y = np.sqrt(R**2-x**2)
y2 = -np.sqrt(R**2-x**2)
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title('Окружность')
plt.grid()
```

```python
a=5
b=3
x = np.linspace(-5, 5, 8000)
y = b/a*np.sqrt(a**2-x**2)
y2 = -b/a*np.sqrt(a**2-x**2)
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
#plt.xlim(-5, 5)
#plt.ylim(-5, 5)
plt.axis('equal')
plt.title('Эллипс')
plt.grid()
```

```python
a=1
b=1
x = np.linspace(-5, 5, 5000)
y = b*(np.sqrt(x**2/a**2-1))
y2 = -b*(np.sqrt(x**2/a**2-1))
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axis('equal')
plt.title('Гипербола')
plt.grid()
```

**5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.**



```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-5, 5, 2)
X, Y = np.meshgrid(X, Y)
Z = 3*X+5*Y+1
Z2 =3*X+5*Y+60
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
plt.title ('График двух параллельных плоскостей')
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-100, 100, 0.1)
Y = np.arange(-100, 100, 0.1)
X, Y = np.meshgrid(X, Y)
Z = 100 * (np.sqrt((X**2)/90+(Y**2)/160))
Z2 = -100 * (np.sqrt((X**2)/90+(Y**2)/160))
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
plt.title ('График конуса второго порядка')
show()
```
