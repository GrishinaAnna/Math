**3. Задание  
Напишите код, который будет переводить полярные координаты в декартовы.**

```python
import math
def poli(r,alfa):
    x = r*math.cos(math.radians(alfa))
    y = r*math.sin(math.radians(alfa))
    return (x, y)
print(poli(4,90))
```

**Напишите код, который будет рисовать график окружности в полярных координатах.**

```python
import numpy as np
import matplotlib.pyplot as plt
alfa = np.linspace(0, 2*np.pi, 360)
r = [5 for i in alfa]
plt.polar(alfa, r, color = 'Red')
plt.show()
```

**4. Задание    
Решите систему уравнений:  
y = x2 – 1  
exp(x) + x∙(1 – y) = 1**


```python
x = np.linspace(-2, 3, 100)
plt.plot(x, (1-(1-np.exp(x))/x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,8) 
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1, y1 =  fsolve(equations, (-1, 2))

x2, y2 = fsolve(equations, (3, 6))
              
print (x1, y1)
print (x2, y2)
```

**Решите систему уравнений и неравенств:  
y = x2 – 1  
exp(x) + x∙(1 – y) > 1**

```python
def equations(p):
    x, y = p
    return (y - x**2 + 1, (np.exp(x) + x - x*y - 1)>0)

x1, y1 =  fsolve(equations, (-1, 2))

x2, y2 = fsolve(equations, (3, 6))
              
print (x1, y1)
print (x2, y2)
```
